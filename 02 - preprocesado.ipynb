## Cleaning Periodo Column

La función cleaning_periodo procesa el DataFrame al descomponer la columna PERIODO, que contiene datos en un formato que combina año y bimestre, en dos nuevas columnas: AÑO y BIMESTRE. Extrae los primeros cuatro caracteres de PERIODO para crear AÑO y el quinto carácter para crear BIMESTRE, ambos convertidos a tipo numérico. Finalmente, elimina la columna original PERIODO, dejando el DataFrame más limpio y estructurado para análisis posteriores.

```python
def cleaning_periodo(data_fixed):
  data_fixed['AÑO'] = pd.to_numeric(data_fixed['PERIODO'].astype(str).str[:4])
  data_fixed['BIMESTRE'] = pd.to_numeric(data_fixed['PERIODO'].astype(str).str[4])
  data_fixed=data_fixed.drop(columns=["PERIODO"])
```

## Cleaning ESTU_PRGM_ACADEMICO column





La función cleaning_programa_academico transforma la columna ESTU_PRGM_ACADEMICO utilizando un codificador de etiquetas (LabelEncoder). Este codificador convierte los valores categóricos de la columna en números enteros, lo que facilita su uso para el análisis y modelado de datos.

## Cleaning ESTU_VALORMATRICULAUNIVERSIDAD column

La función cleaning_matricula procesa la columna ESTU_VALORMATRICULAUNIVERSIDAD del DataFrame al mapear los valores categóricos de matrícula a valores numéricos específicos, definidos en un diccionario. Después de la conversión, calcula la media y la desviación estándar de la columna. Para los valores faltantes (NaN), genera muestras aleatorias de una distribución normal basada en estos parámetros y las asigna a las posiciones correspondientes. Esto permite reemplazar los datos faltantes con valores que mantienen la distribución original, mejorando así la integridad del conjunto de datos.

```python
def cleaning_matricula(data_fixed):
    valores_matricula = {
        'Entre 5.5 millones y menos de 7 millones': 6.25,
        'Entre 2.5 millones y menos de 4 millones': 3.25,
        'Entre 4 millones y menos de 5.5 millones': 4.75,
        'Más de 7 millones': 7.0,
        'Entre 1 millón y menos de 2.5 millones': 1.75,
        'Entre 500 mil y menos de 1 millón': 0.75,
        'Menos de 500 mil': 0.25,
        'No pagó matrícula': 0,
    }
    data_fixed['ESTU_VALORMATRICULAUNIVERSIDAD'] = data_fixed['ESTU_VALORMATRICULAUNIVERSIDAD'].map(valores_matricula)
    mean = data_fixed['ESTU_VALORMATRICULAUNIVERSIDAD'].mean()
    std = data_fixed['ESTU_VALORMATRICULAUNIVERSIDAD'].std()
    sample = np.random.normal(mean, std, data_fixed['ESTU_VALORMATRICULAUNIVERSIDAD'].isna().sum())
    data_fixed.loc[data_fixed['ESTU_VALORMATRICULAUNIVERSIDAD'].isna(), 'ESTU_VALORMATRICULAUNIVERSIDAD'] = sample
```

```python
cleaning_matricula(data_fixed)
data_fixed.head()
```

## Cleaning ESTU_HORASSEMANATRABAJA column



La función cleaning_horas_trabajo transforma la columna ESTU_HORASSEMANATRABAJA del DataFrame al igual que en la función anterior

```python
def cleaning_horas_trabajo(data_fixed):
    horas_promedio = {
        'Menos de 10 horas': 5,
        '0': 0,
        'Más de 30 horas': 35,
        'Entre 21 y 30 horas': (21 + 30) / 2,
        'Entre 11 y 20 horas': (11 + 20) / 2
    }
    data_fixed['ESTU_HORASSEMANATRABAJA'] = data_fixed['ESTU_HORASSEMANATRABAJA'].map(horas_promedio)
    mean = data_fixed['ESTU_HORASSEMANATRABAJA'].mean()
    std = data_fixed['ESTU_HORASSEMANATRABAJA'].std()
    sample = np.random.normal(mean, std, data_fixed['ESTU_HORASSEMANATRABAJA'].isna().sum())
    data_fixed.loc[data_fixed['ESTU_HORASSEMANATRABAJA'].isna(), 'ESTU_HORASSEMANATRABAJA'] = sample
```

## Cleaning FAMI_ESTRATOVIVIENDA column


La función cleaning_estrato procesa la columna FAMI_ESTRATOVIVIENDA del DataFrame  al reemplazar los valores faltantes (NaN) con la moda de la columna, es decir, el valor más frecuente. Después de este llenado, se utiliza un el LabelEncoder para convertir los valores categóricos de FAMI_ESTRATOVIVIENDA en números enteros. Esto simplifica la representación de los estratos de vivienda, facilitando su análisis y modelado en el conjunto de datos.

```python
def cleaning_estrato(data_fixed):
  moda = data_fixed['FAMI_ESTRATOVIVIENDA'].mode()[0]
  data_fixed['FAMI_ESTRATOVIVIENDA'] = data_fixed['FAMI_ESTRATOVIVIENDA'].fillna(moda)
  label_encoder = LabelEncoder()
  data_fixed['FAMI_ESTRATOVIVIENDA'] = label_encoder.fit_transform(data_fixed['FAMI_ESTRATOVIVIENDA'])
```

```python
cleaning_estrato(data_fixed)
data_fixed.head()
```

## Cleaning FAMI_TIENEINTERNET column


La función cleaning_internet transforma la columna FAMI_TIENEINTERNET del DataFrame. Primero, reemplaza los valores faltantes (NaN) con 'Sin Información'. Luego, se utiliza el método map para convertir los valores de la columna en números enteros, asignando 1 para 'Si', 0 para 'No' y -1 para 'Sin Información'. Este proceso facilita el análisis cuantitativo de la disponibilidad de internet en las familias, estandarizando la representación de los datos.

```python
def cleaning_internet(data_fixed):
    data_fixed['FAMI_TIENEINTERNET'] = data_fixed['FAMI_TIENEINTERNET'].fillna('Sin Información')
    data_fixed['FAMI_TIENEINTERNET'] = data_fixed['FAMI_TIENEINTERNET'].map({
        'Si': 1,
        'No': 0,
        'Sin Información': -1
    })
```

```python
cleaning_internet(data_fixed)
data_fixed.head()
```

## Cleaning FAMI_EDUCACIONPADRE column

La función cleaning_padre procesa la columna FAMI_EDUCACIONPADRE del DataFrame. Primero, reemplaza los valores faltantes (NaN) con 'Sin Información' y se asigna -1 a las categorías 'No sabe', 'No Aplica' y 'Sin Información'. Luego, se define un orden específico para los niveles de educación y se aplica un codificador ordinal (OrdinalEncoder) a los valores válidos, convirtiéndolos en enteros según el orden establecido. Finalmente, se convierte la columna a tipo entero, lo que estandariza y facilita el análisis de la educación de los padres en el conjunto de datos.

```python
def cleaning_padre(data_fixed):
    data_fixed['FAMI_EDUCACIONPADRE'] = data_fixed['FAMI_EDUCACIONPADRE'].fillna('Sin Información')
    data_fixed['FAMI_EDUCACIONPADRE'] = data_fixed['FAMI_EDUCACIONPADRE'].replace({
        'No sabe': -1, 'No Aplica': -1, 'Sin Información': -1
    })
    levels = ['Ninguno', 'Primaria incompleta', 'Primaria completa',
              'Secundaria (Bachillerato) incompleta', 'Secundaria (Bachillerato) completa',
              'Técnica o tecnológica incompleta', 'Técnica o tecnológica completa',
              'Educación profesional incompleta', 'Educación profesional completa',
              'Postgrado']
    mask = data_fixed['FAMI_EDUCACIONPADRE'] != -1
    encoder = OrdinalEncoder(categories=[levels])
    data_fixed.loc[mask, 'FAMI_EDUCACIONPADRE'] = encoder.fit_transform(data_fixed.loc[mask, ['FAMI_EDUCACIONPADRE']])
    data_fixed['FAMI_EDUCACIONPADRE'] = data_fixed['FAMI_EDUCACIONPADRE'].astype(int)
```

```python
cleaning_padre(data_fixed)
data_fixed.head()
```

## Converting target column

La función cleaning_rendimiento transforma la columna RENDIMIENTO_GLOBAL de el DataFrame. Donde se utiliza un diccionario de mapeo (rmap) para convertir las categorías de rendimiento académico ('alto', 'medio-alto', 'medio-bajo' y 'bajo') en valores numéricos correspondientes (3, 2, 1 y 0, respectivamente).





```python
def cleaning_rendimiento(data_fixed):
  rmap = {'alto': 3, 'bajo':0, 'medio-bajo':1, 'medio-alto':2}
  data_fixed['RENDIMIENTO_GLOBAL'] = data_fixed['RENDIMIENTO_GLOBAL'].map(rmap)
```

## Function with the complete cleaning

En la siguiente función se hace un llamado a todas la funciones anteriores, esto con el objetivo de facilitar el cleaning general, ya que si posteriormente se cambia el cleaning de alguna de las columnas solo se debe volver a ejecutar la función de dicha columna, y se ejecuta la función general donde las demás funciones están en memoria, de esta manera se evita estar ejecutando todas las funciones una y otra vez por separado.

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np

def data_cleaning(data):
    data_fixed = data.copy()
    cleaning_periodo(data_fixed)
    cleaning_programa_academico(data_fixed)
    cleaning_departamento(data_fixed)
    cleaning_matricula(data_fixed)
    cleaning_horas_trabajo(data_fixed)
    cleaning_estrato(data_fixed)
    cleaning_internet(data_fixed)
    cleaning_padre(data_fixed)
    cleaning_madre(data_fixed)
    cleaning_pago(data_fixed)
    if 'RENDIMIENTO_GLOBAL' in data_fixed.columns:
      cleaning_rendimiento(data_fixed)
    return data_fixed
```
